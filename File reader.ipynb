```python
import os
import io
import sys
from html.parser import HTMLParser


OpenedTags = {}
ClosedTags = {}

class HTMLTagsParser(HTMLParser):
    def handle_starttag(self, tag, attrs):

        if tag not in OpenedTags:

            OpenedTags[tag] = 1

        else:

            OpenedTags[tag] += 1

    def display (self):

        print("\n Opening tags dictionary : {}".format(OpenedTags))
        print("\n Closing tags dictionary : {}".format(ClosedTags))

    def handle_endtag(self, tag):

        if tag not in ClosedTags:

            ClosedTags[tag] = 1

        else:

            ClosedTags[tag] += 1




def insert_line_para_nums(fileRead):
    fileOpen = io.open(fileRead, mode='r', encoding="utf8")
    lineCount = 0
    paragraphCount = 0
    empty = True
    for i in fileOpen:
        if '\n' in i:
            lineCount += 1
            paragraphCount = paragraphCount + 1
    fileOpen.close()
    print("Paragraphs : {} \n".format(paragraphCount))


def parse_file(path):

    fileRead = io.open(path, mode='r',encoding="utf8")
    spaces = tabs = words = numLines = totalChracters = specialCharacter = 0
    linesInFile = fileRead.read()
    totalChracters = len(linesInFile)
    words = len(linesInFile.split())
    numLines = linesInFile.count('\n') + 1
    tabs = len(linesInFile.split('\t')) - 1
    spaces = len(linesInFile.split(' ')) - 1
    specialCharacter = totalChracters - spaces - tabs
    fileRead.close()

    return spaces , tabs , numLines , words , specialCharacter


def choice_function():
    choice = input("1. Input .txt files to know the number of spaces, tabs, lines, words, non-space characters and paragraphs\n2. Input html files to know the numbers of opening and closing tags\n\nEnter your choice : ")

    try:
        value = int(choice)

        if(value == 1):

            fileNames = list(input("Please enter 1 or more text files (format: file1.txt,file2.txt) : ").split(','))
            for fileName in fileNames:
                if os.path.isfile(fileName) is True:

                    spaces , tabs , lines , words , specialCharacter = parse_file(fileName)
                    print("Details of the file {} is given below:".format(fileName))
                    print ("\nSpaces : {} \nTabs : {} \nLines : {} \nWords : {} \nNon-space Characters : {}".format (spaces, tabs, lines, words, specialCharacter))
                    insert_line_para_nums(fileName)
                else :
                    print(fileName, "Not exists")

        elif (value == 2):

            fileNames = list(input("Please enter 1 or more html files (format: file1.html,file2.html) : ").split(','))
            #print(fileNames)
            for fileName in fileNames:
                if os.path.isfile(fileName) is True:
                    parser = HTMLTagsParser()
                    f = open(fileName, 'r')
                    data = f.read()
                    #print(type(data))
                    print("Tags Parsing of file {} :".format(fileName))
                    parser.feed(data)
                    parser.display()
                    OpenedTags ={}
                else :
                    print(fileName, "Not exists")


        else:

            print("That's not an available option !!")

    except ValueError:

        print("That's not an available option !!")


choice_function()
```
